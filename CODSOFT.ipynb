```python
import tkinter as tk
from tkinter import messagebox

def calculate():
    try:
        # Get user input
        num1 = float(entry_num1.get())
        num2 = float(entry_num2.get())
        operation = operation_var.get()

        # Perform calculation
        if operation == '+':
            result = num1 + num2
        elif operation == '-':
            result = num1 - num2
        elif operation == '*':
            result = num1 * num2
        elif operation == '/':
            if num2 != 0:
                result = num1 / num2
            else:
                messagebox.showerror("Error", "Division by zero is not allowed")
                return
        else:
            messagebox.showerror("Error", "Invalid operation")
            return

        # Display result
        label_result.config(text=f"Result: {result:.2f}")

    except ValueError:
        messagebox.showerror("Error", "Invalid input. Please enter valid numbers.")

# Create the main window
root = tk.Tk()
root.title("Simple Calculator")

# Create and place widgets
tk.Label(root, text="Number 1:").grid(row=0, column=0, padx=10, pady=5)
entry_num1 = tk.Entry(root)
entry_num1.grid(row=0, column=1, padx=10, pady=5)

tk.Label(root, text="Number 2:").grid(row=1, column=0, padx=10, pady=5)
entry_num2 = tk.Entry(root)
entry_num2.grid(row=1, column=1, padx=10, pady=5)

tk.Label(root, text="Operation:").grid(row=2, column=0, padx=10, pady=5)
operation_var = tk.StringVar(value='+')
tk.Radiobutton(root, text="+", variable=operation_var, value='+').grid(row=2, column=1, padx=5, pady=5, sticky="w")
tk.Radiobutton(root, text="-", variable=operation_var, value='-').grid(row=2, column=1, padx=5, pady=5)
tk.Radiobutton(root, text="*", variable=operation_var, value='*').grid(row=2, column=1, padx=5, pady=5, sticky="e")
tk.Radiobutton(root, text="/", variable=operation_var, value='/').grid(row=2, column=1, padx=5, pady=5)

tk.Button(root, text="Calculate", command=calculate).grid(row=3, column=0, columnspan=2, pady=10)

label_result = tk.Label(root, text="Result: ")
label_result.grid(row=4, column=0, columnspan=2, pady=10)

# Run the application
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import random
import string

def generate_password():
    try:
        length = int(entry_length.get())
        if length <= 0:
            messagebox.showerror("Error", "Length must be a positive integer.")
            return
        
        # Define character sets
        characters = string.ascii_letters + string.digits + string.punctuation
        
        # Generate password
        password = ''.join(random.choice(characters) for _ in range(length))
        
        # Display the password
        label_password.config(text=f"Generated Password: {password}")
    
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid integer for length.")

# Create the main window
root = tk.Tk()
root.title("Password Generator")

# Create and place widgets
tk.Label(root, text="Password Length:").grid(row=0, column=0, padx=10, pady=10)
entry_length = tk.Entry(root)
entry_length.grid(row=0, column=1, padx=10, pady=10)

tk.Button(root, text="Generate Password", command=generate_password).grid(row=1, column=0, columnspan=2, pady=10)

label_password = tk.Label(root, text="Generated Password: ")
label_password.grid(row=2, column=0, columnspan=2, pady=10)

# Run the application
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import random

# Function to determine the winner
def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (user_choice == "Rock" and computer_choice == "Scissors") or \
         (user_choice == "Scissors" and computer_choice == "Paper") or \
         (user_choice == "Paper" and computer_choice == "Rock"):
        return "You win!"
    else:
        return "You lose!"

# Function to handle button clicks
def play(choice):
    # Computer's choice
    computer_choice = random.choice(["Rock", "Paper", "Scissors"])
    
    # Determine the result
    result = determine_winner(choice, computer_choice)
    
    # Update scores
    if result == "You win!":
        global user_score
        user_score += 1
    elif result == "You lose!":
        global computer_score
        computer_score += 1
    
    # Update labels
    label_computer_choice.config(text=f"Computer chose: {computer_choice}")
    label_result.config(text=result)
    label_user_score.config(text=f"Your Score: {user_score}")
    label_computer_score.config(text=f"Computer Score: {computer_score}")

    # Ask if the user wants to play again
    play_again = messagebox.askyesno("Play Again?", "Do you want to play another round?")
    if play_again:
        label_computer_choice.config(text="Computer chose: ")
        label_result.config(text="")

# Create the main window
root = tk.Tk()
root.title("Rock, Paper, Scissors Game")

# Initialize scores
user_score = 0
computer_score = 0

# Create and place widgets
tk.Label(root, text="Choose Rock, Paper, or Scissors:").grid(row=0, column=0, columnspan=3, padx=10, pady=10)

tk.Button(root, text="Rock", command=lambda: play("Rock")).grid(row=1, column=0, padx=10, pady=10)
tk.Button(root, text="Paper", command=lambda: play("Paper")).grid(row=1, column=1, padx=10, pady=10)
tk.Button(root, text="Scissors", command=lambda: play("Scissors")).grid(row=1, column=2, padx=10, pady=10)

label_computer_choice = tk.Label(root, text="Computer chose: ")
label_computer_choice.grid(row=2, column=0, columnspan=3, pady=10)

label_result = tk.Label(root, text="")
label_result.grid(row=3, column=0, columnspan=3, pady=10)

label_user_score = tk.Label(root, text="Your Score: 0")
label_user_score.grid(row=4, column=0, padx=10, pady=10)

label_computer_score = tk.Label(root, text="Computer Score: 0")
label_computer_score.grid(row=4, column=2, padx=10, pady=10)

# Run the application
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox, simpledialog

# Database to store contacts
contacts = {}

def add_contact():
    name = simpledialog.askstring("Input", "Enter the contact's name:")
    if not name:
        return
    if name in contacts:
        messagebox.showinfo("Info", "Contact already exists.")
        return
    
    phone = simpledialog.askstring("Input", "Enter the phone number:")
    email = simpledialog.askstring("Input", "Enter the email address:")
    address = simpledialog.askstring("Input", "Enter the address:")
    
    if not phone or not email or not address:
        messagebox.showwarning("Warning", "All fields must be filled.")
        return
    
    contacts[name] = {'phone': phone, 'email': email, 'address': address}
    update_contact_list()
    messagebox.showinfo("Info", "Contact added successfully.")

def view_contacts():
    contact_list = "\n".join([f"{name}: {info['phone']}" for name, info in contacts.items()])
    if contact_list:
        messagebox.showinfo("Contact List", contact_list)
    else:
        messagebox.showinfo("Contact List", "No contacts available.")

def search_contact():
    search_term = simpledialog.askstring("Search", "Enter the name or phone number to search:")
    if not search_term:
        return

    results = [f"{name}: {info['phone']}" for name, info in contacts.items() if search_term in name or search_term in info['phone']]
    if results:
        messagebox.showinfo("Search Results", "\n".join(results))
    else:
        messagebox.showinfo("Search Results", "No contacts found.")

def update_contact():
    name = simpledialog.askstring("Input", "Enter the contact's name to update:")
    if not name or name not in contacts:
        messagebox.showwarning("Warning", "Contact not found.")
        return
    
    phone = simpledialog.askstring("Input", "Enter the new phone number:")
    email = simpledialog.askstring("Input", "Enter the new email address:")
    address = simpledialog.askstring("Input", "Enter the new address:")
    
    if not phone or not email or not address:
        messagebox.showwarning("Warning", "All fields must be filled.")
        return
    
    contacts[name] = {'phone': phone, 'email': email, 'address': address}
    update_contact_list()
    messagebox.showinfo("Info", "Contact updated successfully.")

def delete_contact():
    name = simpledialog.askstring("Input", "Enter the contact's name to delete:")
    if not name or name not in contacts:
        messagebox.showwarning("Warning", "Contact not found.")
        return
    
    del contacts[name]
    update_contact_list()
    messagebox.showinfo("Info", "Contact deleted successfully.")

def update_contact_list():
    contact_list = "\n".join([f"{name}: {info['phone']}" for name, info in contacts.items()])
    if contact_list:
        label_contact_list.config(text=contact_list)
    else:
        label_contact_list.config(text="No contacts available.")

# Create the main window
root = tk.Tk()
root.title("Contact Management System")

# Create and place widgets
tk.Button(root, text="Add Contact", command=add_contact).pack(pady=10)
tk.Button(root, text="View Contact List", command=view_contacts).pack(pady=10)
tk.Button(root, text="Search Contact", command=search_contact).pack(pady=10)
tk.Button(root, text="Update Contact", command=update_contact).pack(pady=10)
tk.Button(root, text="Delete Contact", command=delete_contact).pack(pady=10)

label_contact_list = tk.Label(root, text="No contacts available.", justify=tk.LEFT)
label_contact_list.pack(pady=10, padx=10)

# Run the application
root.mainloop()
```
